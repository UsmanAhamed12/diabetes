## Preparing The Dateset

```python
Data = pd.DataFrame(diabetes["data"], columns=diabetes["feature_names"])
Data.head()
```

```python
Data["target"] = diabetes["target"]
Data.head()
```

```python
Data.info()
```

```python
for feature in Data.columns:
    print(f"{feature} has {len(Data[feature].unique())} unique lables")
```

```python
## Summarizing the stats of the data
Data.describe().T
```

```python
## Check the missing values
Data.isnull().sum()
```

```python
## Exploratrry Data Analysis

import seaborn as sns

sns.heatmap(Data.corr(), annot=True)
plt.show()
```

```python
sns.scatterplot(x=Data["bmi"], y=Data["target"], data=Data)
plt.show()
```

```python
sns.regplot(x=Data["bmi"], y=Data["target"], data=Data)
plt.show()
```

```python
sns.regplot(x=Data["s5"], y=Data["target"], data=Data)
plt.show()
```

```python
for feature in Data.columns.drop("target"):
    sns.regplot(x=Data[feature], y=Data["target"], data=Data)

    plt.show()
```

```python
X = Data.drop("target", axis=1)
X.head()
```

```python
y = Data.iloc[:, -1]
```

```python
## Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
## standadize the dataset
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle

pickle.dump(scaler, open("scalling.pkl", "wb"))
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
## print the coffeciatn
model.coef_
```

```python
model.intercept_
```

```python
## On which parameters the model has been trained
model.get_params()
```

```python
## prediction with test data
pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
## plot a scatter plot for the prediction
plt.scatter(y_test, pred)
plt.show()
```

```python
## Residels
residuls = y_test - pred
residuls
```

```python
sns.displot(residuls, kind="kde")
plt.show()
```

```python
## Scatter plot respect to residuls
plt.scatter(pred, residuls)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, median_absolute_error

print(median_absolute_error(y_test, pred))
print(mean_squared_error(y_test, pred))
print(mean_squared_error(y_test, pred) ** 0.5)
print(np.sqrt(mean_squared_error(y_test, pred)))
```

## R Square and Adjusted R Squared 

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, pred)
score
```
